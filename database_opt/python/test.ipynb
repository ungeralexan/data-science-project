```python
from supabase import create_client

supabase = create_client("YOUR_PROJECT_URL", "YOUR_ANON_KEY")

row = {
  "title": "AI Meetup (updated)",
  "source_message_id": "<msg-1@example>"   # same ID as before on purpose
}

# insert or update on conflict
res = supabase.table("test_events").upsert(row, on_conflict="source_message_id").execute()
print("Upserted:", res.data)

# read back to confirm
rows = supabase.table("test_events").select("*").order("created_at", ascending=False).limit(5).execute().data
print("Latest rows:", rows)
```

```python
# pip install supabase
from supabase import create_client

SUPABASE_URL = ""
ANON_KEY = ""
sb = create_client(SUPABASE_URL, ANON_KEY)

rows = [
    {"title": "AI Meetup",         "source_message_id": "<msg-201@example>"},
    {"title": "NLP Colloquium",    "source_message_id": "<msg-202@example>"},
    {"title": "Data Viz Workshop", "source_message_id": "<msg-203@example>"},
]

# insert-or-update in one call; duplicates won’t error
res = sb.table("test_events").upsert(rows, on_conflict="source_message_id").execute()
print("Upserted:", res.data)

# read back to confirm
print(sb.table("test_events").select("*").order("created_at", desc=True).limit(10).execute().data)

```
